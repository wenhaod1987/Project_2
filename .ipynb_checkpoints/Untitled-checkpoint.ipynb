```python
def createplot():
    # coding: utf-8


    # coding: utf-8


    # coding: utf-8

    # # Bokeh Scatterplot v3

    # IMPORT DEPENDENCIES
    import webbrowser
    import pandas as pd 
    import numpy as np 
    import datetime
    from bokeh.plotting import figure, show, output_notebook, output_file
    from bokeh.io import output_file, curdoc, export_png, push_notebook
    from bokeh.models import HoverTool, HTMLTemplateFormatter, CheckboxGroup, DateFormatter, BooleanFilter, Legend, Slider, RedoTool, UndoTool, Panel,Tabs, Button, PointDrawTool, ColumnDataSource, TextInput, Row, ResetTool,NumeralTickFormatter, Column, Select, CustomJS, Tabs, Panel, CDSView, NumberFormatter, Span, Label, GMapPlot, GMapOptions, ColumnDataSource, Circle, LogColorMapper, BasicTicker,DataRange1d, PanTool, WheelZoomTool, BoxSelectTool, Range1d, TapTool, HoverTool, LassoSelectTool, BoxZoomTool
    from bokeh.resources import CDN
    from bokeh.embed import file_html
    from bokeh.application import Application
    from bokeh.application.handlers.function import FunctionHandler
    from bokeh.layouts import gridplot, layout, widgetbox
    from API_Config_Keys import Google_API

    #-----------------------------------------------------------------------------------------------------------------#
    # ACTIONITEM - IMPORT PLOTS FROM OTHER PY FILES AND USE IN SERVER
    #from Bokeh_FLV_Runs import layout as FLV
    # IMPORT ZONDA DATA

    data = pd.read_csv('Redfin_Sample.csv',encoding='UTF-8')
    data.head()
    data.dtypes
```

```python
    df = data
    colors = ['azure','green','firebrick','coral','darkgoldenrod','yellowgreen','lightskyblue','indigo','saddlebrown','navy','orangered','gold','darkgray','blue','bisque','thistle','r','darkgreen','lightsage','darkslategrey','chocolate','cadetblue','mediumvioletred','magenta']
    font = {'family': 'arial','color':  'black','weight': 'normal','size': 16,}
    #-----------------------------------------------------------------------------------------------------------------#
    # FOR LOOP OVER EVERY NEIGHBORHOOD AND GENERATE A NEW RANDOM COLOR FOR EACH NEIGHBORHOOD
    df['Color'] = 'blue'

    x_min = df['SQUARE FEET'].min()
    x_max = df['SQUARE FEET'].max()
    y_min = df.PRICE.min()
    y_max = df.PRICE.max()

    sq_ft_min = TextInput(title="Minimum Home Size (SF)",value=str(x_min))
    sq_ft_max = TextInput(title="Maximum Home Size (SF)",value=str(x_max))
    base_min = TextInput(title="Minimum Base PRICE",value=str(y_min))
    base_max = TextInput(title="Maximum Base PRICE",value=str(y_max))
    # linebutton = Button(label="Test Floorplans",button_type="success")


    # CREATE EMPTY DICTIONARIES FOR COLUMN DATA SOURCES
    source = ColumnDataSource(data=dict(SF=[], PRICE=[], Builder=[], Color=[], Neighborhood=[], LotSize=[], Market=[], Sales_Rate =[], Total_Homes=[], Homes_Sold=[], Unsold_Homes=[], Beds=[], Baths=[], HOA=[], Tax_Rate=[], Floors=[], Lat=[], Long=[], Submarket=[], OpenDate=[]))

    # CREATE FIGURE
    p = figure(x_axis_label ='Home Size (SF)',y_axis_label='Home Price',plot_width=1200, plot_height=800, title="Redfin MLS Listings")
    p.yaxis.formatter=NumeralTickFormatter(format="$ 0,0")
    p.xaxis.formatter=NumeralTickFormatter(format="0,0")


    #PLOT LINES
    #ACTIONITEM CREATE BUTTON TO PLOT LINES
    #p.multi_line(xs="x", ys="y", source=source, size=15, color="Color", line_color='black', fill_alpha=.25)
    p.circle(x="SQUARE FEET", y="PRICE", source=source, size=15, color="blue", line_color='black', fill_alpha=.25)
    r1 = p.circle(x="SQUARE FEET", y="PRICE", source=source, size=15, color="blue", line_color='black', fill_alpha=.25)

    # ADD HORIZONTAL LINES FOR FHA 
    # fha = Span(location=403000,dimension='width', line_color='blue',line_dash='dashed', line_width=.1)


    # FUNCTION TO POPULATE SOURCE DICTIONARIES
    def select_segments():
        selected = df[
            (df['SQUARE FEET'] >= float(sq_ft_min.value)) &
            (df['SQUARE FEET'] <= float(sq_ft_max.value)) &
            (df.PRICE >= float(base_min.value)) &
            (df.PRICE <= float(base_max.value)) 
            ]

        return selected


    def update():
        df = select_segments()
        df = df.sort_values(by=['PRICE'],ascending=True)
    # ACTIONITEM    
        p.title.text = "%d Homes Selected, %d - Average List Price, %d - Average Home Size" % (len(df), df.PRICE.mean(), df['SQUARE FEET'].mean())
        source.data = dict(
            SF=df['SQUARE FEET'],
            Price=df['PRICE']
        )

    callback = CustomJS(args=dict(source=source), code="""
    console.log('tell me this works');
	""")

    TapCallback = CustomJS(args=dict(source=source), code="""
            source.data.do_hover[0] = !source.data.do_hover[0];
        """)

    hover = HoverTool(tooltips = 
        [('Price','@PRICE'),
        ('Home Size', '@SQUARE FEET')
        ],callback = callback,
        show_arrow=False,
        point_policy='snap_to_data')


#     # ASSIGN CALLBACK TO CONTROLS WIDGETS
#     controls = [sq_ft_min, sq_ft_max, base_min, base_max]
#     for control in controls:
#         control.on_change('value', lambda attr, old, new: callback())



    # CREATE LAYOUT 
    inputs = widgetbox([sq_ft_min, sq_ft_max, base_min ,base_max], sizing_mode="fixed")
    cmalayout = gridplot([[p,inputs]], sizing_mode='fixed',merge_tools=False)
    #from Bokeh_FLV_Runs import flvplayout

    # CALL THE UPDATE FUNCTION FOR THE INITIAL LOAD OF DATA
    update()
    curdoc().add_root(cmalayout)
    output_notebook()
    show(cmalayout)

#     return cmalayout

```
